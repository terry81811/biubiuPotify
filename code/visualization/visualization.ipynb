```python
def plotCDF(data, label, feature_name, data_type='review'):
    if data_type == 'review':
        rec_data = sorted(data[label == 1])
        nonRec_data = sorted(data[label == -1])
    elif data_type == 'restaurant' or data_type == 'user':
        rec_data = sorted(data[label == False])
        nonRec_data = sorted(data[label == True])   
    plt.figure()
    plt.suptitle(feature_name)
    plt.axis([-0.01, rec_data[len(rec_data) - 50], 0, 1]) 
    plt.plot(rec_data, np.linspace(0, 1, len(rec_data)), color='green')
    plt.plot(nonRec_data, np.linspace(0, 1, len(nonRec_data)), color='red')
    plt.show()
```

```python
# review feature plot CDF (Pittsburgh)
review_df = pd.read_csv('../../data/review_features_labeled.tsv', delimiter='\t')

label = review_df['label']

df = review_df.drop(['k', 'biz', 'user', 'rating', 'date', 'rank_time_frame', 'label'], 1)

for feature_name in df.columns:
    plotCDF(df[feature_name], label, feature_name, 'review')
```

```python
# restaurant feature plot CDF (Pittsburgh)
rest_df = pd.read_csv('../../data/restaurant_features_labeled.tsv', delimiter='\t')

label = rest_df['fraction'].values > 0.2

df = rest_df.drop(['k', 'numRecR', 'numNonRecR', 'fraction'], 1)

for feature_name in df.columns:
    plotCDF(df[feature_name], label, feature_name, 'restaurant')
```

```python
# user feature plot CDF (Pittsburgh)
user_df = pd.read_csv('../../data/user_features.tsv', delimiter='\t')

# print user_df
label = user_df['fraction'].values > 0.5

df = user_df.drop(['k', 'fraction'], 1)

for feature_name in df.columns:
    plotCDF(df[feature_name], label, feature_name, 'user')
```
